```python
dict_tabs = {}
dict_mdg = {}
dict_not = {}
dict_dec = {}

for file in glob.glob('/content/drive/MyDrive/Colab Notebooks/LMEM-SHAP/All Data Matix2/*.csv'):
  if 'TABS' in(file.split('/')[-1]):
    dict_tabs[int(file.split('/')[-1].split('_')[0][2:])] = file
  if 'DEC' in(file.split('/')[-1]):
    dict_dec[int(file.split('/')[-1].split('_')[0][2:])] = file
  if 'NOT' in(file.split('/')[-1]):
    dict_not[int(file.split('/')[-1].split('_')[0][2:])] = file 
  if 'MDG' in(file.split('/')[-1]):
    dict_mdg[int(file.split('/')[-1].split('_')[0][2:])] = file 
```
